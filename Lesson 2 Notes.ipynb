The above is not really a for loop, you can only have if that that a, and only one expression before the for loop. Its a very simply expression. You cannot create new variable in the expression.

It is list commprehension

[expression for item in iterable if condition]

expression is the value to be added to the new list.
item is the variable that takes the value of each element in the iterable.
iterable is the collection of items to iterate over.
condition is an optional filter that determines whether the item should be included in the new list.
